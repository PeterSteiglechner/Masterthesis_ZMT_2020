```python
import xarray as xr 
import matplotlib.pyplot as plt 
import numpy as np 
import matplotlib as mpl 
from matplotlib.ticker import MaxNLocator
rcParams = {'font.size':20}
plt.rcParams.update(rcParams)
```

```python
folder_std = "/home/peter/EasterIslands/Runs_22May/Standard/"
folders = [folder_std+"FullModel_grid50_gH17e-3_noRegrowth_highFix_linear_NormPop_alphaStd_seed"+str(seed)+"/" for seed in np.arange(1,26)]
stds = [xr.open_dataset(folder+"Statistics.ncdf") for folder in folders]
label_Standard1 = "Standard"
#label_Standard = "Standard: High N fixation\n(without tree regrowth)"
```

```python
from scipy.stats import variation
```

```python
len(stds)
```

```python
x=[]
for u in [5,10,15,20, 25]:
    popData = np.array([stds[k].pop_ct.sum(dim='triangles') for k in range(u)])
    x.append(variation(popData))
```

```python
fig = plt.figure(figsize=(16,9))
ax= fig.add_subplot(111)
plt.plot(stds[0].time,x[0], lw=3, label="5 Runs")
plt.plot(stds[0].time,x[1],  lw=3,label="10 Runs")
plt.plot(stds[0].time,x[2],  lw=3,label="15 Runs")
plt.plot(stds[0].time,x[3],  lw=3,label="20 Runs")
plt.plot(stds[0].time,x[4], lw=3, label="25 Runs")

plt.legend(ncol=5, loc="lower center")
ax.set_title("Standard Run", x=0.01, y=0.9,fontweight="bold", loc='left')
ax.set_ylim(0,)
ax.set_xlim(800,1900)
plt.ylabel("Coefficient of Variation")
plt.xlabel(r"Year [${\rm A.D.}$]")
plt.savefig(folder_std+"CoeffOfVariation.pdf", bbox_inches="tight")
```
